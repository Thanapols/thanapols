```python
##Sphere LS Maximize

import numpy as np

# Create a mathematical function as function (def)
def sphere(x):
    return x[0] ** 2 + x[1] ** 2

# Derivative of f(x)
def sphere_derivative(x):
    return np.array([2 * x[0], 2 * x[1]])

# Local search (LS)
def LS(f, f_derivative, S, step_size, t_max):
    t = 0

    # Find starting point
    x = np.random.uniform(low=S[0], high=S[1], size=(2,))
    print(x)

    # Looping
    while t < t_max:
        y = f(x)
        print("t =", t, ": f(", x, ") =", y)

        # Calculate gradient
        q = f_derivative(x)
        x = x + step_size * q  # Maximize by adding the gradient
        t += 1
    return x, f(x)

x_best, y_best = LS(sphere, sphere_derivative, S=[-15, 15], step_size=0.1, t_max=100)

print("Maximum value of f(x*) =", y_best)
print("Optimal solution x* =", x_best)
```

```python
##Ackley LS Minimize

import numpy as np

def ackley(x):
    return -20 * np.exp(-0.2 * np.sqrt(0.5 * (x[0]**2 + x[1]**2))) + \
            - np.exp(0.5 * (np.cos(2 * np.pi * x[0]) + np.cos(2 * np.pi * x[1]))) + np.e + 20

def ackley_derivative(x):
    factor1 = -20 * np.exp(-0.2 * np.sqrt(0.5 * (x[0]**2 + x[1]**2)))
    factor2 = -0.5 * np.exp(0.5 * (np.cos(2 * np.pi * x[0]) + np.cos(2 * np.pi * x[1])))

    term1 = factor1 * (x[0] / np.sqrt(0.5 * (x[0]**2 + x[1]**2)))
    term2 = factor2 * (-2 * np.pi * np.sin(2 * np.pi * x[0]))

    term3 = factor1 * (x[1] / np.sqrt(0.5 * (x[0]**2 + x[1]**2)))
    term4 = factor2 * (-2 * np.pi * np.sin(2 * np.pi * x[1]))

    return np.array([term1 + term2, term3 + term4])

def LS(f, f_derivative, S, step_size=0.1, t_max=1000, tol=1e-5):
    t = 0

    # Find starting point
    x = np.random.uniform(low=S[0], high=S[1], size=(2,))
    print(x)

    # Looping
    while t < t_max:
        y = f(x)
        print("t =", t, ": f(", x, ") =", y)

        # Calculate gradient
        q = f_derivative(x)
        x = x - step_size * q  # Minimize by subtracting the gradient

        if np.linalg.norm(q) < tol:
            break

        t += 1

    return x, f(x)

# Define the search space
search_space = [-5, 5]

# Perform local search on Ackley function to minimize
min_point, min_value = LS(ackley, ackley_derivative, S=search_space, step_size=0.1)

print("Minimum value of Ackley function:", min_value)
print("Optimal solution:", min_point)
```

```python
import numpy as np
import math
import random

def ackley_function(x, y):
    term1 = -20 * np.exp(-0.2 * np.sqrt(0.5 * (x**2 + y**2)))
    term2 = -np.exp(0.5 * (np.cos(2 * np.pi * x) + np.cos(2 * np.pi * y)))
    result = term1 + term2 + 20 + np.e
    return result

def simulated_annealing_ackley(initial_solution, temperature, cooling_rate, iterations):
    current_solution = initial_solution
    current_energy = ackley_function(*current_solution)

    best_solution = current_solution
    best_energy = current_energy

    for _ in range(iterations):
        # Generate a neighboring solution
        neighbor_solution = current_solution + np.random.uniform(-0.1, 0.1, size=len(current_solution))

        # Clip the values to stay within a reasonable range
        neighbor_solution = np.clip(neighbor_solution, -5, 5)

        # Evaluate the energy of the neighboring solution
        neighbor_energy = ackley_function(*neighbor_solution)

        # Accept the neighbor solution if it's better or with a certain probability
        if neighbor_energy > current_energy or random.random() < math.exp((neighbor_energy - current_energy) / temperature):
            current_solution = neighbor_solution
            current_energy = neighbor_energy

        # Update the best solution if needed
        if current_energy > best_energy:
            best_solution = current_solution
            best_energy = current_energy

        # Reduce the temperature
        temperature *= cooling_rate

    return best_solution, best_energy

# Example usage:
np.random.seed(42)
initial_solution = np.random.uniform(-5, 5, size=2)
temperature = 10.0
cooling_rate = 0.95
iterations = 1000

best_solution, best_energy = simulated_annealing_ackley(initial_solution, temperature, cooling_rate, iterations)

print("Best Solution:", best_solution)
print("Maximum Value of Ackley Function:", best_energy)
```
